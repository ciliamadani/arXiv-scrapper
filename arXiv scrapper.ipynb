```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import PyPDF2 
import os 
```

```python
## goal is to get top 1000 recent 
## so we need to scrape 10 page, each page has 100 article 

list_pdf =[]
for i in range(0,1000,100):
               
    URL = "https://arxiv.org/search/?query=2022&searchtype=all&abstracts=show&order=-announced_date_first&size=100&start="+str(i)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    list_pdf = list_pdf + soup.find_all("a", string="pdf")
    print(i)
```

```python
list_pdf
```

```python
finalPdf=[]
for pdf in list_pdf[:10]:
    finalPdf.append(pdf.attrs['href']) 
```

```python
for i in range(0,len(finalPdf)):
    urllib.request.urlretrieve(finalPdf[i], "C:\\Users\\AzurComputer\\Desktop\\pdf\\pdf"+str(i)+".pdf")
    print("pdf"+str(i)+"downloaded")
```
